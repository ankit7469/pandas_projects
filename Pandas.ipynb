```python
# grouping and aggregation
import pandas as pd
data={
    'Department':['IT','IT','Finance','Finanace','HR'],
    'Salary':[50000,46000,44444,78000,43000]
              }
df=pd.DataFrame(data)
print(df.groupby('Department')['Salary'].mean())
```

```python
# Missing data(Nan handling)
import numpy as np
import pandas as pd
data={
    'Name':['Ankit','aman','Arjun','Abhishek'],
    'Age':[20,np.nan,22,23],
    'Marks':[55,np.nan,55,45]}
df=pd.DataFrame(data)  
print(df)
print(df.isnull())                # check nan
print(df.fillna(0))               # fill na with 0
print(df.dropna())                # removes nan rows
```

```python
# sorting
import pandas as pd
data={
    'Name':['Ankit','aman','Arjun','Abhishek'],
    'Age':[20,21,22,23],
    'Marks':[55,60,55,45]}
df=pd.DataFrame(data)  
print(df.sort_values(by='Marks'))
```

```python
# merge,join and concat
df1=pd.DataFrame({'ID':[1,2],'Name':['Ankit','Abhishek']})
df2=pd.DataFrame({'ID':[1,2],'marks':['55','45']})

# merge
print(pd.merge(df1,df2,on='ID'))

# concat
print(pd.concat([df1,df2],axis=0))
```
